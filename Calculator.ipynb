```python
print("*****Calculator*****")
print("")

# Menu
print("1.Addition")
print("2.Subtraction")
print("3.Multiplication")
print("4.Division")
print("5.Exit")
print("-----------")
print("")


#user input the two numbers
num1=int(input("Enter Number 1 :"))
num2=int(input("Enter Number 2 :"))

print("")

print("-----------")

#user inputs the choice
choice=int(input("Choose the option :"))

print("")
print("************")

#process according to the choice & gets the answer

if choice==1:
    answer=num1+num2
    print("Your Answer is :" ,answer)
elif choice==2:
    answer=num1-num2
    print("Your Answer is :" ,answer)
elif choice==3:
    answer=num1*num2
    print("Your Answer is :" ,answer)
elif choice==4:
    answer=num1/num2
    print("Your Answer is :" ,answer)
elif choice==5:
    print("Good Bye !")
```
